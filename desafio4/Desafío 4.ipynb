```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


HRDataset_v14 = pd.read_csv('./HRDataset_v14.csv')
fig, ax = plt.subplots(figsize = ( 5 , 3 )) 
ax.set_xlabel( "Salarios en Dolares" , size = 12 ) 
ax.set_ylabel( "Cantidad de empleados" , size = 12 ) 
ax.set_title( "Salarios mas Populares" , size = 24 ) 
sns.histplot(data = HRDataset_v14, x = "Salary", kde = True)
plt.show()
ax = sns.catplot(data = HRDataset_v14, kind="violin" , x="Sex" , y = "Salary", split=True)
ax.set(xlabel='Sexo', ylabel='Salarios', title='Salarios, Discriminados por Género')
plt.show()
sns.lineplot(data = HRDataset_v14,x ='DateofHire')

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as stats

HRDataset_v14 = pd.read_csv('./HRDataset_v14.csv')

#Mediana
print('La mediana de lo salarios es:',stats.median(HRDataset_v14["Salary"]))

#Media
print('La media de los salarios es:', np.mean(HRDataset_v14["Salary"]))

#Moda
print("La moda de los salarios es:",(stats.mode(HRDataset_v14["Salary"])))



fig, ax = plt.subplots(figsize = ( 5 , 3 )) 
ax.set_xlabel( "Salarios en Dolares" , size = 12 ) 
ax.set_ylabel( "Cantidad de empleados" , size = 12 ) 
ax.set_title( "Salarios mas Populares" , size = 24 ) 
sns.histplot(data = HRDataset_v14, x = "Salary", kde = True)
plt.show()

ax = sns.catplot(data = HRDataset_v14, kind="violin" , x="Sex" , y = "Salary", split=True)
ax.set(xlabel='Sexo', ylabel='Salarios', title='Salarios, Discriminados por Género')
plt.show()


serieTiempo = pd.read_csv('./HRDataset_v14.csv', parse_dates=['DateofHire'], index_col='DateofHire').sort_index()

def plot_df(df, x, y, title="", xlabel='Fecha', ylabel='Salarios', dpi=100):
    plt.figure(figsize=(16,5), dpi=dpi)
    plt.plot(x, y, color='tab:red')
    plt.gca().set(title=title, xlabel=xlabel, ylabel=ylabel)
    plt.show()
    
plot_df(serieTiempo, x=serieTiempo.index, y=serieTiempo.Salary, title='Evolucion de los salarios a travez del tiempo') 


```

# Hacer un array aliatorio de 100 numeros

```python
import numpy as np
x = np.random.rand(100)*1000
x = x.astype(int)
print(x)
```

# Calcular la mediana

```python
mediana = np.median(x)
print(mediana)
```

# Calcular factorial

```python
def facto():
     val = input("Indique el número: ") or 1  
     return np.math.factorial(int(val))
print (facto())
```

# suma los números que indique el usuario entre un número inicial y uno final

```python
def sumatoria():
    while True:  
     inicial = int(input("Indique el número inicial: ") or 1)  
     final = int(input("Indique el número final: ") or 100)  
     if inicial<final:  
      break  
    x = np.array(range(inicial,final+1))
    suma = np.sum(x)
    return print("Suma de los números comprendidos entre ",inicial," y ",final,"= ",suma)  
sumatoria()
```

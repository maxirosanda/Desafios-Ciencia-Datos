```python
#Carga de las librerías
import pandas as pd
import seaborn as sns
import urllib.request
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('bank-additional-full.csv',sep=";")
```

Numéricas:

```python
data = data.drop(["housing"], axis=1)
```

```python
data = data.drop(["month"], axis=1)
```

```python
data = data.drop(["day_of_week"], axis=1)
```

```python
data = data.drop(["age"], axis=1)
```

```python
data = data.drop(["loan"], axis=1)
```

```python
data = data.drop(["duration"], axis=1)
```

```python
data = data.drop(["campaign"], axis=1)
```

```python
data = data.drop(["pdays"], axis=1)
```

```python
data = data.drop(["emp.var.rate"], axis=1)
```

```python
data = data.drop(["cons.price.idx"], axis=1)
```

```python
data = data.drop(["cons.conf.idx"], axis=1)
```

```python
data = data.drop(["nr.employed"], axis=1)
```

```python
data = data.drop(["default"], axis=1)
```

```python
#Separamos en X e y
X = data.drop(["y"], axis=1)
y = data["y"]
```

```python
#Separamos en train y test!
(X_train, X_test,
y_train, y_test) = train_test_split(X,
                                    y,
                                    stratify=y,
                                    test_size=0.30,
                                    random_state=11)
```

```python
#Creamos un arbol de decisión sencillo y lo fiteamos
tree = DecisionTreeClassifier(random_state=11)
tree.fit(X_train, y_train)
```

```python
y_test_pred = tree.predict(X_test) #Prediccion en Test
```

```python
from sklearn.metrics import accuracy_score

#Calculo el accuracy en Test
test_accuracy = accuracy_score(y_test, y_test_pred)

print('% de aciertos sobre el set de evaluación:',test_accuracy)
```

```python
#Creamos un random forest!
model = RandomForestClassifier(random_state=11, n_estimators=200,
                               class_weight="balanced", max_features="log2")
model.fit(X_train, y_train)
```

```python
y_test_pred = model.predict(X_test) #Prediccion en Test
```

```python
#Calculo el accuracy en Test
test_accuracy = accuracy_score(y_test, y_test_pred)

print('% de aciertos sobre el set de evaluación:',test_accuracy)
```

Conclusiones!

Simplemente utilizando el algoritmo de random forest, obtenemos un mejor accuracy que haciendo uso de un DecisionTreeClassifier 😉

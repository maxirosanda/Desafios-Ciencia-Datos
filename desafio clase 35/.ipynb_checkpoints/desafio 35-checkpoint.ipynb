```python
#Importamos las librerias
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
#Importemos el archivo
df = pd.read_csv('./bank-full.csv',sep=";")
```

```python
#Principales medidas estadisticas
df.describe().T
```

```python
sns.scatterplot(data=df, x="age", y="duration")
```

```python
#se observa que las personas mas jovenes mantienen una conversacion mas larga
```

```python
sns.boxplot(x=df['default'], y=df['duration'])
```

```python
#en las personas con default se observa mas cantidad de Outliers
```

```python
cat = df[["job", "duration"]].groupby(['job'], as_index=False).sum().sort_values(by='duration', ascending=False)

plt.figure(figsize=(20,8))

sns.barplot(x='job', y='duration', data=cat)
```

```python
#se obserban conversaciones mas largas en personas con trabajos del area de blue-collar ,management
#,technician,admin y services
```

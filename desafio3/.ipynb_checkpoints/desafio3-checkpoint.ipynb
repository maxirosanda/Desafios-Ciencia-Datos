```python
import pandas as pd
import numpy as np
desafio3 = pd.read_csv('./Summer-Olympic-medals-1976-to-2008.csv')
year_datetime = pd.to_datetime(desafio3['Year'],format ='%Y%')
print(year_datetime)
desafio3 = desafio3.set_index("Year")
apellido_nombre = desafio3["Athlete"].str.split(",",expand=True)
desafio3["Apellido"] = apellido_nombre[0]
desafio3["Nombre"] = apellido_nombre[1]
print(desafio3)
paises_cant_medallas = desafio3["Country"].value_counts()
pais_mas_medallas = paises_cant_medallas.idxmax()
print("El pais con mas medallas es " + pais_mas_medallas)
medallas_hombre_año = desafio3[desafio3.Gender == 'Men']
medallas_hombre_año =medallas_hombre_año.groupby(['Year'])["Gender"].value_counts()
print(medallas_hombre_año)
```

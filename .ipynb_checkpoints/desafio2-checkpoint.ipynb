# Calcular factorial

```python
def facto():
     val = input("Indique el número: ") or 1  
     return np.math.factorial(int(val))
print (facto())
```

# suma los números que indique el usuario entre un número inicial y uno final

```python
def sumatoria():
    while True:  
     inicial = int(input("Indique el número inicial: ") or 1)  
     final = int(input("Indique el número final: ") or 100)  
     if inicial<final:  
      break  
    x = np.array(range(inicial,final+1))
    suma = np.sum(x)
    return print("Suma de los números comprendidos entre ",inicial," y ",final,"= ",suma)  
sumatoria()
```

```python
import pandas as pd
File = pd.ExcelFile('Salarios_Openqube.xlsx')
df=File.parse('datos')
salarios_brutos =  df.loc[:, 'Salario mensual NETO (en tu moneda local)']
salarios_brutos = np.array(salarios_brutos)
salarios_brutos = [int(x) for x in salarios_brutos]
salarios_media =np.mean(salarios_brutos)
salarios_mediana =  np.median(salarios_brutos)
print("media de los salarios:") 
print(salarios_media)
print("mediana de los salarios")
print(salarios_mediana)
```

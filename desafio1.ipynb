# calculo de factorial

```python

def facto():
    from math import factorial
    val = input("Indique el número: ") or 1
    val = factorial(int(val))
    return val
facto()
```

# suma los números que indique el usuario entre un número inicial y uno final  

```python
def sumatoria():
    while True:  
     inicial = int(input("Indique el número inicial: ") or 1)  
     final = int(input("Indique el número final: ") or 100)  
     if inicial<final:  
      break  
    suma=0  
    for i in range(inicial,final+1):  
     suma+=i  
    return print("Suma de los números comprendidos entre ",inicial," y ",final,"= ",suma)  
sumatoria()
```
